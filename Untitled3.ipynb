## Making An IMDB Predictive dataset for positive and negative reviews

```python
# importing libraries
import os
import pandas as pd
import numpy as np

# Importing the preprocessing libraries
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,classification_report

# Deep learning framework
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from  tensorflow.keras.layers import Embedding,Dense,LSTM,Dense, Dropout
import nltk
nltk.download('stopwords')

```

```python
df= pd.read_csv('/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv')
```

```python
def load_data(data_dir):
  reviews,labels = [],[]
  for label_type in ['pos','neg']:
    dir_name = os.path.join(data_dir,label_type)
    for fname in os.listdir(dir_name):
      if fname.endswith('.txt'):
        with open(os.path.join(dir_name, fname),encoding = 'UTF-8') as f:
          reviews.append(f.read())
          labels.append( 1 if label_type == 'pos' else 0)
          return reviews, labels

train_dir = 'path_to _your_train_data'
test_dir = 'path_to_your_test_data'

train_reviews, train_labels = df
test_reviews, test_labels = df
```

## Cleaning the text by removing stopwords,lowercasing and tokenizing


```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
# since df is the only dataframe containing my values
train_reviews = df ['review'].values
train_labels = df ['sentiment'].values

# Split data into training and testing data
train_reviews, test_reviews,train_labels, test_labels = train_test_split(
    train_reviews,train_labels, test_size= 0.2 , random_state=42)


stop_words = stopwords.words('english')

def preprocess_text(reviews):
  vectorizer = TfidfVectorizer(stop_words=stop_words, max_features=10000)
  X = vectorizer.fit_transform(reviews)
  return X

x_train = preprocess_text(train_reviews)
x_test = preprocess_text(test_reviews)
y_train,y_test = np.array(train_labels),np.array(test_labels)
```

```python
df.isna()
```

## Training a Model(using Naives Bayes classifier)

```python
model = MultinomialNB()
model.fit(x_train,y_train)

# Predicting on test set
y_pred = model.predict(x_test)

# Evaluating the model
print("Accuracy:",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred,target_names=["Negative","Positive"]))
```

```python
# To better visualize the model
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report

# Calculate the classification report
# convert to lowercase
#y_pred = np.char.lower(y_pred)

# Check the data types and convert if necessary
if not np.issubdtype(y_test.dtype, np.number):
  # Assuming your labels are -Ve and +ve
  y_test = np.where(y_test == 'positive', 1,0)
if not np.issubdtype(y_pred.dtype,np.number):
  y_pred = np.where(y_pred == 'positive',1,0)

report = classification_report(y_test, y_pred,target_names =["Negative","Positive"],output_dict=True)

# convert the report to a pandas dataframe
import pandas as pd
report_df = pd.DataFrame(report).transpose()


# Create the heatmap using Seaborn
plt.figure(figsize=(8,6))
sns.heatmap(report_df.iloc[:-1, :-1],annot=True, cmap='Blues')
plt.title('classification Report Heatmap')
plt.show()
```

## For a more Advanced deep learning approach using LSTM


```python
# Tokenize the data
max_words = 10000
max_len = 500

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(train_reviews)

x_train_seq = tokenizer.texts_to_sequences(train_reviews)
x_test_seq = tokenizer.texts_to_sequences(test_reviews)

x_train_pad = pad_sequences(x_train_seq, maxlen=max_len)
x_test_pad = pad_sequences(x_test_seq,maxlen=max_len)
```

```python
## Building the LSTM model
model = Sequential()
model.add(Embedding(max_words,128,input_length=max_len))
model.add(LSTM(128,dropout=0.2, recurrent_dropout=0.2)) # Regularization methods
model.add(Dense(1,activation='sigmoid'))

model.compile(loss= 'binary_crossentropy',
              optimizer= 'adam',
              metrics=['accuracy'])
## converting labels to numerical values(0 and 1)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)


# Training the model
model.fit(x_train_pad,y_train,batch_size=64,epochs=5,validation_split= 0.2)

```

```python
## Evaluating model

# converting labels to numerical values (0 and 1)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_test = le.fit_transform(y_test) # preprocess y_test

score,acc =model.evaluate(x_test_pad,y_test)
print("Test accuracy:", acc)

#Checking for the classification report(precision,recall,F1-score)
y_pred_dl = model.predict(x_test_pad)
y_pred_dl = np.round(y_pred_dl).astype(int)

print(classification_report(y_test,y_pred_dl,target_names=["Negative","Positive"]))
```
